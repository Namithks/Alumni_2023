```python
pairs = [
    ["(.*)Your Name(.*)", ["We have the names of all alumni in our database."]],
    ["(.*)Phone Number(.*)", ["Sure, we have contact details of the alumni."]],
    ["(.*)Batch(.*)", ["Yes, we have information about the batches of alumni."]],
    ["(.*)Your Permanent Address(.*)", ["We have permanent addresses of the alumni."]],
]
```

```python
chatbot = Chat(pairs, reflections)
```

```python
app = Flask(__name__)

@app.route("/")
def home():
    return render_template("index.html")

@app.route("/get", methods=["POST"])
def get_bot_response():
    user_message = request.form["msg"]
    return str(chatbot.respond(user_message))

if __name__ == "__main__":
    app.run(debug=True)
```
